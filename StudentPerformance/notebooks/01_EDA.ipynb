```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the dataset
df = pd.read_csv(r'C:\Users\thili\DataAnalyticsPortfolio_T\StudentPerformance\data\StudentsPerformance.csv')

# Show Structure of the dataset
print(df.shape)
print(df.head())

# Summary
print(df.describe())
print(df.dtypes)

# Plot score destribution
plt.figure(figsize=(10, 6))
df[['math score', 'reading score', 'writing score']].hist(bins=20)
plt.suptitle("Score Distributions")
plt.tight_layout()
plt.show()

# Gender-wise score distribution
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
sns.boxplot(x='gender', y='average score', data=df)
plt.title('Average Score by Gender')
plt.show()
```
